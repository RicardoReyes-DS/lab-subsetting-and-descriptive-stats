#### Hint: Use string functions

```python
# your answer here

# Reference: https://www.geeksforgeeks.org/python-pandas-series-str-find/
orders_df['LB_Orders'] = orders_df['Description'].str.find('lunch bag')

lb_ = orders_df[orders_df.LB_Orders != -1]
lb_
```
